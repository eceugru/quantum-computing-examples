```python
from qiskit import *
from qiskit.visualization import plot_histogram
from qiskit.visualization import plot_bloch_multivector
%matplotlib inline
import math
```

```python
circuit= QuantumCircuit(2,2)
plot_bloch_multivector(circuit)
```

```python
circuit.h(0)
plot_bloch_multivector(circuit)
```

```python
circuit1= QuantumCircuit(2,2)
circuit1.h(0)
circuit1.cx(0,1)
plot_bloch_multivector(circuit1)
```

Durum vektr kayboldu.nk ikiside entenglement durumunda yani birbirine bal (ya 00 olacak ya da 11) bu yzden Q-sphere ile bakmak daha iyi olacaktr.

```python
circuit_2= QuantumCircuit(2,2)
circuit_2.rx(math.pi/4,0) #ilk q-bit pi/4 kadar rotasyona urat
circuit_2.rx(math.pi/2,1)
plot_bloch_multivector(circuit_2)
```

```python
circuit_2= QuantumCircuit(2,2)
circuit_2.ry(math.pi/4,0) #ilk q-bit pi/4 kadar rotasyona urat
circuit_2.ry(math.pi/2,1)
plot_bloch_multivector(circuit_2)
```
