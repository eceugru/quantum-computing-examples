```python
from qiskit import *
from qiskit.visualization import plot_histogram
%matplotlib inline
```

```python
circuit=QuantumCircuit(2,1)
```

```python
circuit.h(0)
circuit.x(1)
circuit.h(1)
circuit.barrier()
circuit.draw(output='mpl')
```

```python
circuit.cx(0,1)
circuit.barrier()
circuit.h(0)
circuit.barrier()
circuit.draw(output='mpl')
```

```python
circuit.measure(0,0)
```

```python
circuit.draw(output='mpl')
```
